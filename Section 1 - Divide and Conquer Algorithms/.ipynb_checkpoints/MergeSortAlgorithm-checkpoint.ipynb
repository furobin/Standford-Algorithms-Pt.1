# Merge Sort Algorithm

Very old algorithm, used back in 1945. Works well and still used all the time in practice. Standard sorting algo in many programming libraries. Also good introduction to Divide&Conquer (DnC) algorithm paradigm - break problem down into smaller sub problems to solve recursively, then combine and get original solution. Merge-Sort is very clear demonstration of this.

Improves over selection, insertion, and bubble sorts. <br/>

Selection - do number of passes through the data repeatedly, identifying minimum elements that haven't been looked at yet, linear number of passes doing minimal computation each time<br/>

Insertion - repeatedly maintain the invariant that the prefix view of the array which is sorted version of elements, go through and insert remainder back into this prefix <br/>

Bubble - identify adjacent pairs of elements which are out of order, do repeated swaps until end of array is sorted. <br/>
These are mostly all quadratic dependence, O^2

Motivates guiding principles for algorithm analysis with worst-case and asymptotic analysis. Analysis generalizes to Master Method, also called "Recursion-Tree" method here. 

#### The Sorting Problem - Given array of n numbers, unsorted. For output, have a sorted array in increasing order. 

Generally considers the numbers to be distinct, but also consider what happens if there are ties(duplicates).

```python
array = [5,4,1,8,7,2,6,3]
#Split array into two smaller arrays, 1st half to x, 2nd half to y
x = [5,4,1,8]
y = [7,2,6,3] 
#Each array sorts itself w/recursive calls, becomes:
x = [1,4,5,8]
y = [2,3,6,7]
#Merges two arrays, basically walk pointers down both arrays comparing values; traverse through lists in parallel
final = [1,2,3,4,5,6,7,8]
```

```python
def MergeSort(array):
    if len(array) == 0 or len(array) == 1:
        return array
    else:
        x = array[0:len(array)//2]
        y = array[len(array)//2:]
        MergeSort(x)
        MergeSort(y)

        i = j = k = 0
        while i<len(x) and j<len(y):
            if x[i] < y[j]:
                array[k] = x[i]
                i += 1
            else:
                array[k] = y[j]
                j += 1
            k += 1
            
        while i<len(x):
            array[k] = x[i]
            i += 1
            k += 1
        
        while j<len(y):
            array[k] = y[j]
            j += 1
            k += 1

if __name__ == '__main__':
    array = [5,4,1,8,7,2,6,3]
    MergeSort(array)
    print(array)
```

Guiding Principles for Analysis of Algorithms

1. Worst-case analysis - make no assumptions about the input, each stage of analysis assume worst-case analysis. Running time bound holds for every input of length n. Appropriate or general-purpose routines. 

Alternatives:
 - Average-case analysis - avg running time of algo under assumps about relative frequencies of diff inputs
 - Benchmarks - set agreed benchmark inputs that are practical/typical
 Useful but need domain knowledge about problem; not applied here

2. Do not pay uch attention to constant factors, lower-order terms. Think Big-O notation. 

Reasons:
 - Easier to work with mathematically
 - Stresing over constant factors is lame af. Difficult to count exact operations w/loop increments and others. Small constant factors may also differ based on programming languages and on machine-language code itself.
 - Lose very little predictive power. Lose a little exact info but not important 

In practice and industry, constants w/crucial programs will be important. For us here, dw about it

3. Asymptotic Analysis - focus on running time for large imputs. For example, our nlogn is considered better than n^2 even if coefficients and constants are different bc we are focused on large input sizes.

Reasons:
 - Only large input sizes are really the only interesting problems. Small problems pretty much barely differ in run-time even if efficiencies are different. 

Fast algorithms are ones where the worst-case running time grows slowly with input size. Sweet spot: mathematical tractability and predictive power. Holy grail is a linear time algorithm. Won't always be able to achieve, but nlogn (slightly superlinear) is pretty close.
